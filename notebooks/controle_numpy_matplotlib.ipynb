# Partie A — Importation du Dataset, Arrays, Premières Manipulations

### A1. Importation du dataset

# Partie D — Visualisation avec Matplotlib (Plots Simples)

### D17. Line plot
Line plot des longueurs des sépales pour les 50 premières fleurs.


```python
from matplotlib.lines import lineStyles
#Setosa
sepal_length = features[:50, 0]
plt.figure(figsize=(10, 6))
plt.plot(sepal_length, marker = '*' , linestyle='dotted' , color='purple')
plt.xticks(range(0,51,5))
plt.title("Longueur des sepales pour les 50 premieres fleurs(Setosa)")
plt.xlabel("Indice du fleur")
plt.ylabel("Longueur des sepales (cm)")
plt.grid(True)
plt.show()
```

### D18. Scatter plot
Scatter plot : longueur vs largeur des pétales.


```python
petal_length = features[:, 2]
petal_width = features[:, 3]
plt.figure(figsize=(10, 6))
plt.scatter(petal_length, petal_width, c='purple', alpha = 0.7, edgecolors='black')
plt.title("Longueur vs Largueur des petales")
plt.xlabel("Longueur des petales (cm)")
plt.ylabel("Largeur des petales (cm)")
plt.grid(True , alpha=0.4)
plt.show()
```
