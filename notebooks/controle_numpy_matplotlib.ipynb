### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
variance_setosa = features[0:50].var(axis=0).mean()
variance_versicolor = features[50:100].var(axis=0).mean()
variance_virginica = features[100:150].var(axis=0).mean()

# Affichage
print("Variance globale moyenne par espèce :")
print("Setosa :", variance_setosa.round(3))
print("Versicolor :", variance_versicolor.round(3))
print("Virginica :", variance_virginica.round(3))

# L'espèce avec la plus grande variance
if variance_virginica > variance_versicolor and variance_virginica > variance_setosa:
    espece_max = 'Virginica'
elif variance_versicolor > variance_setosa:
    espece_max = 'Versicolor'
else:
    espece_max = 'Setosa'

print("\nEspèce avec la plus grande variance globale :", espece_max)

if espece_max == 'Virginica':
    data = features[100:150]
elif espece_max == 'Versicolor':
    data = features[50:100]
else:
    data = features[0:50]

plt.figure(figsize=(10, 6))
plt.boxplot([data[:, 0], data[:, 1], data[:, 2], data[:, 3]],
            labels=['Sepal Len', 'Sepal Wid', 'Petal Len', 'Petal Wid'])
plt.title(f'Distributions des caractéristiques pour {espece_max}')
plt.ylabel('Valeur (cm)')
plt.show()
```
